```python
from flask import Flask, render_template, request, jsonify
from flask_ngrok import run_with_ngrok
from keras.applications.xception import preprocess_input, decode_predictions
import tensorflow as tf
from keras.preprocessing import image
import numpy as np
from keras.models import load_model
import cv2
from skimage.segmentation import mark_boundaries
import os
import base64
from io import BytesIO
from PIL import Image
```

```python

app = Flask(__name__)

#run_with_ngrok(app)

# Load the trained model
model = load_model('/content/drive/My Drive/checkpoints/Xception/model3_e88_26/model_epoch_10_val_accuracy_1.0000.h5')

@app.route('/predict/',methods=['POST'])
def predict():
    # get JSON data format
    data = request.get_json()

    # Obtener el string en base64 de la imagen
    img_string = data['image']

    # Decodificar el string en base64 a bytes
    img_data = base64.b64decode(img_string)

    # Convertir los bytes a un objeto de imagen PIL
    img = Image.open(BytesIO(img_data))

    #new_size = (224, 224)
    #resized_img = img.resize(new_size)
    # Convert PIL image to NumPy
    img_array = np.expand_dims(img, axis=0)
    img_array = img_array / 255.0

    # Get the model's prediction
    prediction = model.predict(img_array)
    predicted_class = 'Real' if prediction[0][0] > 0.5 else 'Fake'
    print(f'Prediction: {prediction[0][0]}')
    print(f'Predicted class: {predicted_class}')

    return jsonify({'Predicted class': predicted_class})
```
